# MergeSort Сортировка слиянием

## Полный код

```python
def merge_sort(list):
    # 1. Сохраняем длину списка
    list_length = len(list)
    
    # 2. Список меньше длиной, чем уже отсортирован
    if list_length == 1:
        return list
    
    # 3. Определяем середину списка и делим его на две части
    mid_point = list_length // 2
    
    # 4. Убеждаемся, что все разделы разбиты на отдельные компоненты
    # вызывается функция merge_sort, и в качестве параметра передается разделенная часть списка
    left_partition = merge_sort(list[:mid_point])
    right_partition = merge_sort(list[mid_point:])
    
    # 5. Функция merge_sort возвращает отсортированный список левой и правой части
    return merge(left_partition, right_partition)

# 6. Принимаем два списка и возвращаем отсортированный список, состоящий из содержимого в двух списках
def merge(left, right):
    
    # 7. Инициализируем вывод пустого списка, который будет заполнен отсортированными элементами
    # Инициализируем две переменные i и j, которые используются указателями при переборе списков
    output = []
    i = j = 0
    
    # 8. Цикл while выполняется, если оба указателя i и j меньше длины левого и правого списков 
    while i < len(left) and j < len(right):
        
        # 9. Сравнием элемент каждой позиции обоих списков во время итерации
        if left[i] < right[j]:
            # output заполняется наименьшим значением
            output.append(left[i])
            
            # 10. Перемещаем указатель вправо
            i += 1
        else:
            output.append(right[j])
            j += 1
            
    # 11. Оставшиеся элементы выбираются из текущего значения указателя в конец соответствующего списка
    output.extend(left[i:])
    output.extend(right[j:])

    return output

def run_merge_sort():
    unsorted_list = [4, 1, 5, 7, 2, 6, 1, 1, 6, 4, 10, 33, 5, 7, 23]
    print(unsorted_list)
    sorted_list = merge_sort(unsorted_list)
    print(sorted_list)


run_merge_sort()
```
